# Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import GridSearchCV
```

```python
ridge=Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,50,80,90,1e2,1e3,1e5,1e6,1e7,1e10]}
ridge_regressor=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X_train,Y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
rregressor=Ridge(alpha=100)
rregressor.fit(X_train, Y_train)
```

```python
print(rregressor.coef_)
```

```python
Ridgeprediction = rregressor.predict(X_test)
```

```python
RidgeTrain = rregressor.predict(X_train)
```

```python
plt.scatter(Y_test,Ridgeprediction)
```

```python
print('Ridge Regression MAE:', mean_absolute_error(Y_test, Ridgeprediction))
print('Ridge Regression MSE:', mean_squared_error(Y_test, Ridgeprediction))
print('Ridge Regression RMSE:', np.sqrt(mean_squared_error(Y_test, Ridgeprediction)))
```

```python
print('R-Squared value: %.15f' % r2_score(Y_test, Ridgeprediction))
```

```python
print('R-Squared value: %.15f' % r2_score(Y_train, RidgeTrain))
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
```

```python
lassoreg = Lasso()
parameters = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1e-1,1,5,10,20,25,30,35,40,50,1e2,1e3,1e5,1e6,1e7,1e10]}
lasso_regressor = GridSearchCV(lassoreg, parameters, scoring='neg_mean_squared_error', cv=5)

lasso_regressor.fit(X_train,Y_train)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
ll = Lasso(alpha = 1)
ll.fit(X_train, Y_train)
```

```python
print(ll.coef_)
```

```python
Lassoprediction = ll.predict(X_test)
```

```python
LassoTrain = ll.predict(X_train)
```

```python
plt.scatter(Y_test,Lassoprediction)
```

```python
print('Lasso Regression MAE:', mean_absolute_error(Y_test, Lassoprediction))
print('Lasso Regression MSE:', mean_squared_error(Y_test, Lassoprediction))
print('Lasso Regression RMSE:', np.sqrt(mean_squared_error(Y_test, Lassoprediction)))
```

```python
print('R-Squared value: %.15f' % r2_score(Y_test, Lassoprediction))
```

```python
print('R-Squared value: %.15f' % r2_score(Y_train, LassoTrain))
```
